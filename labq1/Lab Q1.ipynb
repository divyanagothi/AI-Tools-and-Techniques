# Food profit predict

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd         #to work on the datasets
```

```python
#reading data
data=pd.read_csv("profitdataset.txt")
print(data.shape)
```

```python
# get x and y data values
x=data['population'].values
y=data[' profit'].values
```

```python
#plot the data values
plt.scatter(x,y,40,c='g',label='scatter_data',marker='o')
plt.xlabel('population')
plt.ylabel(' profit')
plt.title('food_truck_profit_graph')
plt.grid(True,color='k')
plt.show()
```

```python
# mean values
mean_x=np.mean(x)
mean_y=np.mean(y)
n=len(x)
mean_x
```

```python
#y=mx+c. to get the values m and c
num=0
den=0
for i in range(n):
    num=num+((x[i]-mean_x)*(y[i]-mean_y))
    den=den+((x[i]-mean_x)**2)
m=num/den
#c=y-m*x
c=mean_y-(m*mean_x)
print(m,c)
```

```python
#calculate y_predicate
y_pred=np.array([])       #to create empty array

for i in range(n):
    p=(m*x[i])+c           #y=mx+c
    y_pred=np.append(y_pred,np.array([p]),axis=0)
     
print(y_pred)   
```

```python
#plot the regression line
plt.plot(x,y_pred,color='m',label="regression line")
#plot the actual data
plt.scatter(x,y,40,c='c',label="actual_data",marker='*')
plt.scatter(x,y,40,c='k',label="predicted_data",marker='x')
plt.xlabel("population")
plt.ylabel(" profit")
plt.legend()
plt.show()
```

```python
# how good our fit line or regression line is using r square mean

nu=0
de=0
for i in range(n):
    nu=nu+((y_pred[i]-mean_y)**2)    #predicted values of y
    de=de+((y[i]-mean_y)**2)         # actual values of y
r2=nu/de
r2



# r square ranges from 0 to 1 if >0.7. then its good fit line
```
